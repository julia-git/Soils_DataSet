# CLEANING DATA

```python
#REPLACE MISSPELLED WORD "Topk" WITH "Top"
df.replace({'Topk': 'Top'}, inplace=True)
df.head()
```

```python
#Drop nulls
df['pH'].dropna(inplace=True)
df['pH'] #Notice that the null records in index 1 and index 18 are gone. 
```

```python
#Let's bring back the dataset again so we can work with the two null records
df = pd.read_csv("Soils.csv") 
df['pH'].head(20)
```

```python
#Impute null values with mean
df['pH'].fillna(df['pH'].mean(), inplace=True)
df['pH']
```

```python
#Drop column 'Na'
df.drop(columns = ['Na'], inplace = True)
df.head() #Notice that the 'Na' column is gone
```

```python
#Drop row index 2
df.drop(2, axis=0, inplace=True)
df
```

```python
df.rename(columns = {'Conduc' : 'Cond', 'Dens' : 'Density'}, inplace = True)
df.head()
```

```python
df['Cond'].apply(np.sqrt)
```

# DATA AGGREGATION

```python
#Mean of records in "Ca" column, grouped by column "Contour"
df.groupby(by=['Contour'])['Ca'].mean()
```

```python
#Sum of records in "Ca" column, grouped by column "Contour"
df.groupby(by=['Contour'])['Ca'].sum()
```

```python
#Count of records in "Ca" column, grouped by column "Contour"
df.groupby(by=['Contour'])['Ca'].count()
```

```python
#Mean of "Ca" column, grouped by two columns: "Contour" and "Gp"
df.groupby(by=['Contour', 'Gp'])['Ca'].mean()
```

# COMBINING MULTIPLE DATFRAMES

```python
df2 = pd.read_csv("Soils2.csv")
```

```python
##Concatenate data by column
pd.concat([df, df2], axis=1)
```

```python
##Concatenate data by row
pd.concat([df, df2], axis=0)
```

```python
#Outer join on the two dataframes on the "Contour" columns
pd.merge(df, df2, left_on='Contour', right_on='Contour', how='outer')
```
